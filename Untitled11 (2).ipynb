**Read Image**

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread("./1.png")
cv2_imshow(img)
```

**Define Filters**

```python
import numpy as np

def gaussian_kernel(size, sigma=1):
    size = int(size) // 2
    x, y = np.mgrid[-size:size+1, -size:size+1]
    normal = 1 / (2.0 * np.pi * sigma**2)
    g =  np.exp(-((x**2 + y**2) / (2.0*sigma**2))) * normal
    return g

def blur_kernel(size):
      kernel = np.ones((size, size), np.float32)/18

      return kernel
```

```python

def filters(image,kernel_1,kernel_2):

    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    filtered_1 = cv2.filter2D(gray, -1, kernel_1)
    filtered_2 = cv2.filter2D(filtered_1, -1, kernel_2)
    ret, thresh = cv2.threshold(filtered_2, 250, 255, cv2.THRESH_OTSU)
    return thresh

def detect_contour(img, image_shape):

    img_cp = np.zeros(image_shape, np.uint8)
    contours, hierarchy = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    cnt = sorted(contours, key=cv2.contourArea, reverse=True)[0]
    return img_cp, cnt

    cv2.drawContours(img_cp, cnt, -1, (0, 255, 255), 3)
    plt.title('Largest Contour')
    plt.imshow(img_cp)
    plt.show()


def detect_corners(canvas, cnt):

    epsilon = 0.06 * cv2.arcLength(cnt, True)
    approx_corners = cv2.approxPolyDP(cnt, epsilon, True)
    cv2.drawContours(canvas, approx_corners, -1, (255, 255, 0), 10)
    approx_corners = sorted(np.concatenate(approx_corners).tolist())
    approx_corners = [approx_corners[i] for i in [0, 2, 1, 3]]
    return approx_corners


def destination(corners):


    w1 = np.sqrt((corners[0][0] - corners[1][0]) ** 2 + (corners[0][1] - corners[1][1]) ** 2)
    w2 = np.sqrt((corners[2][0] - corners[3][0]) ** 2 + (corners[2][1] - corners[3][1]) ** 2)
    w = max(int(w1), int(w2))

    h1 = np.sqrt((corners[0][0] - corners[2][0]) ** 2 + (corners[0][1] - corners[2][1]) ** 2)
    h2 = np.sqrt((corners[1][0] - corners[3][0]) ** 2 + (corners[1][1] - corners[3][1]) ** 2)
    h = max(int(h1), int(h2))

    destination = np.float32([(0, 0), (w - 1, 0), (0, h - 1), (w - 1, h - 1)])
    return destination, h, w

def unwarp(img, src, dst):
    h, w = img.shape[:2]
    H, _ = cv2.findHomography(src, dst, method=cv2.RANSAC, ransacReprojThreshold=3.0)
    un_warped = cv2.warpPerspective(img, H, (w, h), flags=cv2.INTER_LINEAR)



    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))
    ax1.imshow(img)
    ax1.set_title('Befor')

    x = [src[0][0], src[2][0], src[3][0], src[1][0], src[0][0]]
    y = [src[0][1], src[2][1], src[3][1], src[1][1], src[0][1]]

    ax2.imshow(img)
    ax2.plot(x, y, color='yellow', linewidth=1)
    ax2.set_ylim([h, 0])
    ax2.set_xlim([0, w])
    ax2.set_title('Contour')


    return un_warped

```

```python
    kernel_1 = gaussian_kernel(3, sigma=1)
    kernel_2 = blur_kernel(3)
    filtered_image = filters(img,kernel_1,kernel_2)
    img_cp, largest_contour = detect_contour(filtered_image, img.shape)
    corners = detect_corners(img_cp, largest_contour)
    destination_points, h, w = destination(corners)
    un_warped = unwarp(img, np.float32(corners), destination_points)
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))
    ax1.imshow(un_warped)


```
